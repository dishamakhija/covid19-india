```python
import pandas as pd
import numpy as np
```

```python
##Data Source (https://www.kaggle.com/sudalairajkumar/covid19-in-italy)
```

```python
filepath= "../data/covid19-in-italy/covid19_italy_region.csv"
```

```python
italy_df = pd.read_csv(filepath, sep=",")
```

```python
italy_df.head()
```

```python
##format date
italy_df["date_formatted"] = pd.to_datetime(italy_df["Date"], infer_datetime_format=True).apply(lambda x: x.strftime("%m/%d/%Y"))
```

```python
columns = ["RegionName", "Latitude", "Longitude", "TotalPositiveCases", "date_formatted"]
italy_df = italy_df[columns]
```

```python
italy_df["Country/Region"]="Italy"
```

```python
datewise_df = pd.pivot_table(italy_df, index=['RegionName','Country/Region', 'Latitude', 'Longitude'],
                    columns=['date_formatted'], values=["TotalPositiveCases"], aggfunc=np.sum).fillna(0)
```

```python
datewise_df.columns = [element[1] for element in datewise_df.columns]
```

```python
datewise_df.reset_index(inplace=True)
```

```python
datewise_df.rename(columns={"Latitude": "Lat", "Longitude": "Long", "RegionName": "Province/State"}, inplace=True)
```

```python
datewise_df.to_csv("../data/timeseries_covid19_italy_nhu_confirmed_cases_04102020.csv", index=False)
```
