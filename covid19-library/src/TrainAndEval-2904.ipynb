```python
resultDict = dict()

for idx, expt in experiments.iterrows():
    
    # Set max_evals in sample_train_config and sample_finaltrain_config appropriately
    # Currently set at 10
    sample_train_config = json.loads('{ "region_name": "bengaluru", "region_type": "district", "train_start_date": "4/11/20", "train_end_date": "4/20/20", "model_class": "SEIR", "model_parameters": { "incubation_period": 5}, "search_space": { "r0": [ 0.5, 6 ], "EbyCRatio": [ 0.5, 70 ], "IbyCRatio": [ 0.8, 1 ], "infectious_period": [1,10] }, "search_parameters": { "max_evals": 10 }, "training_loss_function": { "metric_name": "mape", "variable_weights": [ { "variable": "confirmed", "weight": 1 } ] }, "loss_functions": [ { "metric_name": "mape", "variable_weights": [ { "variable": "confirmed", "weight": 1 } ] }, { "metric_name": "rmsle", "variable_weights": [ { "variable": "confirmed", "weight": 1 } ] } ], "output_filepath": "seir_train.json" }')
    sample_test_config = json.loads('{ "region_name": "bengaluru", "region_type": "district", "run_day": "4/20/20", "test_start_date": "4/21/20", "test_end_date": "4/25/20", "model_class": "SEIR", "model_parameters": { "infectious_period": 9.5412, "incubation_period": 5, "IbyCRatio": 0.9, "r0": 2.4091, "LatentEbyCRatio": { "3/22/20": 19.692374309340906, "4/1/20": 0.5781344247805147 } }, "loss_functions": [ { "metric_name": "mape", "variable_weights": [ { "variable": "confirmed", "weight": 1 } ] }, { "metric_name": "rmsle", "variable_weights": [ { "variable": "confirmed", "weight": 1 } ] } ], "output_filepath": "seir_evaluation_output.json" }')
    sample_finaltrain_config = json.loads('{ "region_name": "bengaluru", "region_type": "district", "train_start_date": "4/16/20", "train_end_date": "4/25/20", "model_class": "SEIR", "model_parameters": { "incubation_period": 5, "r0": 2.4091 }, "search_space": { "r0": [ 0.5, 6 ], "EbyCRatio": [ 0.5, 70 ], "IbyCRatio": [ 0.8, 1 ], "infectious_period": [1,10] }, "search_parameters": { "max_evals": 10 }, "training_loss_function": { "metric_name": "mape", "variable_weights": [ { "variable": "confirmed", "weight": 1 } ] }, "loss_functions": [ { "metric_name": "mape", "variable_weights": [ { "variable": "confirmed", "weight": 1 } ] }, { "metric_name": "rmsle", "variable_weights": [ { "variable": "confirmed", "weight": 1 } ] } ], "output_filepath": "seir_train.json" }')


    region = expt['region']
    resultDict[region] = dict()
   
    train_config = sample_train_config.copy()
    train_config['region_name'] = expt['region']
    train_config['region_type'] = expt['region_type']
    train_config['train_start_date'] = expt['train1_start_date']
    train_config['train_end_date'] = expt['train1_end_date']
    
    train_module_config = TrainingModuleConfig.parse_obj(train_config)
    trainResults = TrainingModule.from_config(train_module_config)
    resultDict[region]['Train1MAPE'] = trainResults['train_metric_results'][0]['value']
    resultDict[region]['Train1RMLSE'] = trainResults['train_metric_results'][1]['value']
    resultDict[region]['Train1BestParams'] = trainResults['best_params']
    resultDict[region]['Train1LatentParams'] = trainResults['latent_params']
    print("Training Complete")

    #TODO: Might need to change IbyCRatio if the seed data is available for April 20th
    
    test_config = sample_test_config.copy()
    test_config['region_name'] = expt['region']
    test_config['region_type'] = expt['region_type']
    test_config['test_start_date'] =expt['test1_start_date']
    test_config['test_end_date'] =expt['test1_end_date']
    test_config['run_day'] = expt['run_day']
    test_config['model_parameters'].update(trainResults['best_params'])
    test_config['model_parameters'].update(trainResults['latent_params'])
    test_module_config = ModelEvaluatorConfig.parse_obj(test_config) 
    evalResults = ModelEvaluator.from_config(test_module_config)
    resultDict[region]['Test1MAPE'] = evalResults[0]['value']
    resultDict[region]['Test1RMLSE'] = evalResults[1]['value']
    
    print("Evaluation Complete:", resultDict[region]['Test1MAPE'], resultDict[region]['Test1RMLSE'])
    finalTrain_config = sample_finaltrain_config.copy()
    
    #TODO: Might need to change IbyCRatio if the seed data is available for April 20th
    
    finalTrain_config['region_name'] = expt['region']
    finalTrain_config['region_type'] = expt['region_type']
    finalTrain_config['train_start_date'] =expt['train2_start_date']
    finalTrain_config['train_end_date'] =expt['train2_end_date'] 
    finalTrain_config['output_filepath'] = region + '_output_train.json'

    finalTrain_module_config = TrainingModuleConfig.parse_obj(finalTrain_config)
    finalResults = TrainingModule.from_config(finalTrain_module_config)
    resultDict[region]['Train2MAPE'] = finalResults['train_metric_results'][0]['value']
    resultDict[region]['Train2RMLSE'] = finalResults['train_metric_results'][1]['value']
    resultDict[region]['Train2BestParams'] = finalResults['best_params']
    resultDict[region]['Train2LatentParams'] = finalResults['latent_params']
       
    print(resultDict)
    
    #expt['Train1MAPE'] = trainResults['train_metric_results'][0]['value']
    #expt['Train1BestParams'] =trainResults['best_params']
    #expt['TestMAPE'] = evalResults[0]['value']
    #expt['Train2MAPE'] =finalResults['train_metric_results'][0]['value']
    #expt['Train2BestParams'] = finalResults['best_params']
```
